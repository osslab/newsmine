````python
import glob

files = glob.glob('./newsdata/2017/*/newsisall.json')
````

````python
import json
def read_json_file(path_to_file):
    with open(path_to_file) as p:
        rows=[json.loads(line) for line in p]
    return rows
````

````python
import pandas as pd

data=[]
for path in files:
    data.extend(read_json_file(path))

df = pd.DataFrame(data)
````

````python
df=df.dropna(subset=['section'])
df=df.reset_index(drop=True)
````

````python
import re
def remove_extras(article):
    #return_article = ''
    article = re.sub(r'\/\/[a-zA-Z가-힣 \n]+|{}|_flash_removeCallback\(\)', '', article)
    article = article.replace('▶','\n▶').replace('\t', '').replace('\n', ' ')
    #remove by line
    article = re.sub(r'\【[a-zA-Z가-힣·\(\)\- ]+\=[a-zA-Z가-힣\/ ]+\】[a-zA-Z가-힣·\/ ]+ =|\【[a-zA-Z가-힣·\(\)\/\- ]+\=[a-zA-Z가-힣\/ ]+\】|\【[a-zA-Z가-힣·\(\)\/\- ]+\】|\<< [a-zA-Z가-힣· ]+\=[a-zA-Z가-힣· ]+ >>|\([a-zA-Z가-힣0-9·, ]+\=[a-zA-Z가-힣· ]+\)|\[[a-zA-Z가-힣·= ]+[자료사진 ]\]|\(연합뉴스 DB\)|\[[a-zA-Z가-힣 ]+제공\]|\[[a-zA-Z가-힣 ]+제공=연합뉴스\]', '', article)
    #remove copyrithgs
    #article = article.replace('<저작권자(c) 연합뉴스, 무단 전재-재배포 금지>', '')
    #remove email
    #article = re.sub(r'[a-zA-Z0-9_-]+@[a-z]+.[a-z]+.[a-z]+', '', article)
    #remove tail
    
    #for sentence in article.split('\n'):
    #    if not sentence.startswith('▶') : return_article += sentence

    #return_article = return_article.replace('\n', ' ')
    article = re.sub(r'(20[0-9 ]+\.[0-9 ]+\.[0-9 ]+\.)(.*)', '', article)
    article = re.sub(r'([a-z.0-9]+@newsis.com)(.*)', '', article)

    return article
````
