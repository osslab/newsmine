```python
from pymongo import MongoClient
client = MongoClient('newsdb', 27017)
db = client.news
collections = db['articles']
```

```python
result = collections.find({'date':'2017-06-25'})
```

```python
import pandas as pd

df =  pd.DataFrame(list(result))
```

```python
import string
translator = str.maketrans('', '', string.punctuation)
df['title_re']=df['title'].str.replace('[\xa0|\`|\’|\”|\‘|\“|\"|\'|\[|\]|\.|\,|\·]', ' ')
df
```

```python
import requests
import json
URL = 'http://open-korean-text.herokuapp.com/extractPhrases'


def get_pharase(text):
    params = {'text': text}
    res = requests.get(URL, params=params)
    response = json.loads(res.text)  
    phrases = response["phrases"]
    return_Phrases = []
    for phrase in phrases :
        return_Phrases.append(phrase.split('(')[0])
    
    return return_Phrases
```

```python
get_pharase("포토  6·25전쟁 67주년  기념사하는 이낙연 총리")
```

```python
documents = df['title_re'].tolist()
documents
```

```python
import numpy as np;
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(tokenizer=get_pharase)
vect_weights = vec.fit_transform(documents)
print(vect_weights)
```

```python
weights = np.asarray(vect_weights.mean(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': vec.get_feature_names(), 'weight': weights})
```

```python
weights_df.sort_values(by='weight', ascending=False).head(20)
```

```python
def noun_count (str):
  words_len = len(str.split())
  return words_len
```

```python
weights_df['term_count']=weights_df['term'].apply(noun_count)
phrase_weights_df=weights_df[weights_df['term_count']>2]
```
