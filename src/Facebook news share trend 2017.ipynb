```python
import glob

files = glob.glob('./newsdata/newsreveal/*.json')
```

```python
import json
def read_json_file(path_to_file):
    with open(path_to_file) as p:
        rows=[json.loads(line) for line in p]
    return rows
```

```python
import pandas as pd

data=[]
for path in files:
    data.extend(read_json_file(path))    
    

df = pd.DataFrame(data)
```

```python
len(df)
```

```python
df.drop(['_id','author', 'naver_section'], axis = 1, inplace = True)
```

```python
df['year'] = df['date'].apply(lambda x: x[0] )
df['month'] = df['date'].apply(lambda x: x[1] )
df['total'] = df['shares'].apply(lambda x: x[0] )
```

```python
df
```

```python
df_2017 = df[df['year']=='2017']
df_2016 = df[df['year']=='2016']
```

```python
df_2017.reset_index(drop=True)
```

```python
group_total = df_2017['total'].groupby([df_2017['month'], df_2017['media']]).mean()
```

```python
unstacked_total = group_total.unstack()
unstacked_total.reset_index(level=0, inplace=True)
```

```python
unstacked_total
```

```python
grouped=df_2017.groupby(pd.Grouper(key='month'))
group_sample=grouped.media.value_counts()
```

```python
unstacked=group_sample.unstack(level=-1)
unstacked.sort_index(inplace=True)
unstacked.reset_index(level=0, inplace=True)
unstacked
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()
```

```python
p1 = figure(plot_width=750, plot_height=500)
p1.grid.grid_line_alpha=0.3
p1.xaxis.axis_label = 'Date'
p1.yaxis.axis_label = 'Count'
p1.line(unstacked_total['month'], unstacked_total['chosun'], color='#fba330', legend='조선일보',line_width=2)
p1.line(unstacked_total['month'], unstacked_total['donga'], color='#e86d25', legend='동아일보',line_width=2)
p1.line(unstacked_total['month'], unstacked_total['hani'], color='#2e297c', legend='한겨레',line_width=2)
p1.line(unstacked_total['month'], unstacked_total['jtbc'], color='#cc1f29', legend='JTBC',line_width=2)

show(p1)
```
